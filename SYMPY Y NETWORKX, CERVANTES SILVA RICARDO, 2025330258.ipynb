<h1 style="color:#A82323; font-family:lobster; text-align:center; font-size:200%">Sympy</h1>
<p style="font-family: arial; text-align:justify; font-size:150%">SymPy es una librería de Python diseñada para realizar matemáticas simbólicas. Permite trabajar con expresiones algebraicas de forma exacta (no numérica), como simplificar ecuaciones, derivar, integrar, resolver ecuaciones, trabajar con matrices simbólicas y manipular fórmulas matemáticas. Es muy útil en áreas como álgebra, cálculo, ecuaciones diferenciales y teoría matemática.</p>
<p style="font-family: arial; text-align:justify; font-size:150%">A continuación, se mostrará un código que servirá como ejemplo de la implementación de esta libreria, realizando algunas actividades en el área del álgebra lineal, por ejemplo: multiplicación de vectores con matrices, potencias de matrices, solución de sistemas de ecuaciónes, así como el cálculo de vectores y valores propios.</p>
<p style="font-family: arial; text-align:justify; font-size:150%">Para finalizar con esta sección, se explicarán algunas de las líneas mas significativas del código, para tener un mejor entendimiento del mismo.</p>

```python
import sympy as sm
#Multiplicación de una matriz por un vector y potencia de una matríz
matriz = sm.Matrix([[1,2,3],[9, 6, 2], [4, 2, 5]])
vector = sm.Matrix([3,4,5])
multiplicacion = matriz * vector
m_potencia = matriz**3

#Sistemas de ecuaciónes (A *x = b)
matrizA = sm.Matrix([[2,3,1],[6, -2, -1], [3, 1, -1]])
b = sm.Matrix([1,-14,1])
sol = matrizA.solve(b)

#Vectores y valores propios
valores_p = matriz.eigenvals()
vectores_p = matriz.eigenvects()
print(matriz, "\n", multiplicacion, "\n", m_potencia)
print("\n", sol, "\n")
```

<p style="font-family: arial; text-align:justify; font-size:150%">En este código podemos observar que las operaciones se realizan de forma muy sencilla utilizando esta librería, pero lo complicado puede venir al momento de entender qué es lo que genera cada uno de los comandos, para quitarle un poco lo complicado, se explicarán algunas líneas:
<ul style="font-family: arial; text-align:justify; font-size:150%">
    <li>"sm.Matrix([ [ ], [ ], ... ])": Esta línea nos permite crear una matriz, de tamaño n*m, donde n y m pueden ser iguales o diferentes, dependiendo de lo que queramos hacer; dentro de los corchetes debemos de ingresar las filas de nuestra matriz en orden. Cabe recalcar que, si quieremos crear un vector, debemos de utilizar el mismo comando, solo que únicamente agregaremos una fila.</li>
    <li>"matriz * vector": Representa la multiplicación entre una matriz y un vector, lo que devuelve es un vector (mstriz de una fila) con las entradas correspondientes de la multiplicación.</li>
    <li>"matriz**3": Con esta linea, podemos obtener la potencia n (en este caso 3) de una matriz, como resoltado obtendremos una matriz con las entradas correspondientes a esta operacion, podemos utilizar distintos números de potencia, para el ejemplo se utilizó la potencia 3.</li>
    <li>"matrizA.solve(b)": Con esta línea, lo que acemos es resolver el sistema de ecuaciones de la forma A * x = b, esto es un método del objeto matriz, por lo que accedemos a el con un punto después del nomre de nuestra matriz y el nombre "solve( )" poniendo en los parentesis nuestro vetor de resultados; tendremos como salida, un vector con la solución correspondiente.</li>
    <li>matriz.eigenvals()", "matriz.eigenvects()": Con estas líneas obtendremos los eigenvalores y eigenvectores de nuestra matriz.
    <ul style="font-family: arial; text-align:justify; font-size:90%">
        <li>Eigenvectores: Obtendremos como salida un diccionario, en el que la primer componente es el eigenvalor, y la segunda es su multiplicidad.</li>
        <li>Eigenvalores: Obtendremos una lista de tuplas, la primer componente de la tupla es es valor propio asociado al eigenvector y su multiplicidad, como segunda componente tenemos el eigenvector correspondiente.</li>
    </ul></li>
</ul></p>

<h1 style="color:#A82323; font-family:lobster; text-align:center; font-size:200%">NetworkX</h1>
<p style="font-family: arial; text-align:justify; font-size:150%">NetworkX es una librería de Python utilizada para crear, analizar y visualizar grafos o redes. Permite modelar estructuras formadas por nodos y conexiones (aristas), como redes sociales, rutas de transporte o redes de computadoras. Incluye herramientas para calcular caminos más cortos, centralidad de nodos, comunidades y muchas otras propiedades de redes complejas.</p>
<p style="font-family: arial; text-align:justify; font-size:150%">Igualmente que en la sección anterios, se mostrará un código que servirá como ejemplo de la implementación de esta libreria, específicamente para graficar un diagrama de transicion asociado a una cadena de Markov a tiempo discreto.}</p>
<p style="font-family: arial; text-align:justify; font-size:150%">Similarmente, para finaliza con esta sección, se explicarán algunas de las líneas mas significativas del código, para tener un mejor entendimiento del mismo.</p>

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
transiciones = [
    ("Estado A", "Estado A", 0.8),
    ("Estado A", "Estado B", 0.2),
    ("Estado B", "Estado A", 0.3),
    ("Estado B", "Estado B", 0.7)
]
for origen, destino, prob in transiciones:
    G.add_edge(origen, destino, weight=prob)
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, arrowsize=20)

labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

plt.show()
```

<p style="font-family: arial; text-align:justify; font-size:150%">Es importante mencionar que, en este ejemplo únicamente utilizamos dos estados, sin embargo, pueden ser muchos más; también, utilizamos un grafo dirigido, ya que necesitamos las fechas que nos indiquen cuál es el orden del cambio de estados. Explicaremos algunas líneas de este código.
<ul style="font-family: arial; text-align:justify; font-size:150%">
    <li>"G = nx.DiGraph()": Define el grafo dirigido, que es lo que necesitamos para visualizar las flechas.</li>
    <li>"G.add_edge(origen, destino, weight=prob)": Esta línea define los nodos (representando los estados), y con weight se muestra el valor de la probabilidad para mostrar en la línea.</li>
    <li>"pos = nx.spring_layout(G)": Calcula las coordenadas de los nodos, esto para qye no se amontonen y el diagrama sea legible.</li>
    <li>"nx.draw()", "nx.draw_networkx_edge_labels()": Se renderizan los nodos y las flechas, además, es el comando que hace que podemos escribir sobre la flecha del grafo, para visualizar el valor de la probabilidad.</li>
</ul></p>
