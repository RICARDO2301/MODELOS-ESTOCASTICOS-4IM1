<p style="font-family: arial; text-align:justify; font-size:150%">Tenemos un punto que se puede mover a la derecha con propabilidad \(p\) y a la izquierda con probabilidad \(q\), tal que \(p + q =1 \)</p>
<p style="font-family: arial; text-align:justify; font-size:150%">Para explicar lo que pasa con este problema, se plantea al siguiente código en python, el cual, dado un valor para \(p\), realiza una simulación con 5000 unidades de tiempo para ver como, dependiendo del valor de \(p\), la caminata tiene distintos comportamientos.</p>
<p style="font-family: arial; text-align:justify; font-size:150%">Se invita al lector a modificar el valor de \(p\) para visualizar las diferentes formas que puede tomar la caminata. Para que no existan errores teoricos en cuanto a la probabilidad, se añadieron distintas reestricciones sobre el valor de \(p\).</p>

```python
import random as rd
import matplotlib.pyplot as plt
def caminata(p):
     q=1-p
     num=rd.random()
     if num<=p:
        return 1
     else:
        return -1
pos=[]
n=5000
p=0.499
cont=0
alj=[]
if p<0 or p>1:
        print("No es posible que p tome ese valor")
else:
    for i in range (n+1):
       pos.append(cont)
       alj.append(abs(cont))
       cont+=caminata(p)
    plt.plot(pos)
    plt.suptitle(f"Representación gráfica de la caminata, p={p}")
    plt.axhline(y=0, color='r', linestyle='-')
    plt.xlabel("Unidades de tiempo")
    plt.ylabel("Posición")
    regreso=sum(1 for x in pos if x==0) - 1
    if regreso>0:
       tiem=pos[1:].index(0) +1
       print(f"Después de {n} movimientos, se está en el puesto {cont}\n")
       print(f"Se regresó a la posición inicial {regreso} veces\n")
       print(f"El alejamiento máximo es: {max(alj)}\n")
       print(f"En promedio se aleja del origen {sum(alj)/n}\n")
       print(f"Se tarda {tiem} unidades de tiempo en regresar\n")
    else:
       print("No regresó al origen\n")
```

<p style="font-family: arial; text-align:justify; font-size:150%">Para la simulación "predeterminada" en el código, se propone un valor de \(p = 0.499\), ya que, con este valor podemos notar que hay fluctuaciones en los valores que toma la caminata, esto nos permite hacer uso de las demás funcionalidades implementadas en el código. </p>

<b><p style="color:#D25353; font-family:lobster; text-align:center; font-size:180%">Análisis y Conclusiones</p></b>

<p style="font-family: arial; text-align:justify; font-size:150%">Es imporante precisar que, conforme se cambie el valor de probabilidad de \(p\) (y, por consecuente, el de \(q\)), las caminatas tomarán formas diferentes, por ejemplo, si \(p>0.5\), la caminata se irá, por mucho, a los valores positivos; por el contrario, si \(p<0.5\), la caminatá se irá hacia los valores negativos; además, en caso de que \(p = q\), se podría esperar que la caminata no esté muy lejana del origen, tieniendo valores tanto positivos como negativos, pero no muy alejados del cero. <br>Con el objetivo de profundizar aún mas en la caminata, dentro del código, se analizaron algunos aspectos de ella, por ejemplo:<br> 
<ul style="font-family: arial; text-align:justify; font-size:150%">
    <li>¿El punto regresa al origen?: Para dar respuesta a este cuestionamiento, a través de la lista que creamos para guardar cada una de las posiciones del punto, identificamos si hay alguna igual a cero (excepto la primera, porque sabemos que siempre será cero), en caso de que si, sabemos que volvió al menos una vez al origen. Para complementar este punto, decidí hacer el conteo general de las veces que regresó al origen en todo el recorrido.</li>
    <li>Si regresa, ¿En cuánto tiempo lo hace por primera vez?: Para saber en qué momento regresó al origen por primera vez, lo que hice fue identificar (y guardar) el índice de la lista en el cual se encuentra el primer cero (sin contar el primero), de esta forma, en caso de que regresara, sabemos en qué momento lo hizo.</li>
    <li>¿Cuánto se aleja, en promedio, del origen?: Para esta pregunta hice uso de una formula; primero, sea \[A = \{\, |X_n| \mid n \in \mathbb{N}, \, n\leq5000\}\] donde \(X_n\) es la posicion de la caminata en el momento \(n\); con esto, la formula utilizada es la siguiente: \[\overline{A} = \frac{1}{5000} \sum_{n=1}^{5000} X_n , X_n \in A\] de ahí podemos obtener, en promedio cuando se alejó del origen, utilizamos los valores absolutos porque, como hablamos de distancia, esta siempre es positiva.</li>
    <li>¿Cuál es su alejamiento máximo?: Aquí, sea \(M=max(A)\), sabemos que \(M\) será el alejamiento máximo, ya que, como estamos considerando los valores absolutos para el conjunto \(A\), no importa si el valor es positivo o negativo, se está tomando su distancia hacia el origen y, de esta forma, obtenemos el alejamiento máximo. </li>
    <li>¿Este proceso es una cadena de Markov?: Si, si lo es, porque si analizamos lo que pasa en cada uno de los \(X_n\), notaremos que si valor únicamente dependerá del \(X_{n-1}\), o sea, depende únicamente del estado anterior, y, de esta forma, cumple con la propiedad de Markov, generando que, efectivamente, sea una cadena de Markov.</li>
    <li>¿Es posible saber la distribución de probabilidad de la caminata? Si es posible, inicialmente podríamos tomar unicamente los pasos hacia la derecha, cada uno de ellos seria un experimento de Bernoulii, y, como su naturaleza es sumarlos uno a uno, podriamos pensar en una distribución Binomial con parametros \(n\) y \(p\); para hacerlo computacionalmente, podriamos pensar en el método de Monte Carlo, el cuál consiste en repetir la simulación una gran cantidad de veces para lograr obtener una distribución empírica a través de ellas.</li></ul> </p>
