<p style="font-family: arial; text-align:justify; font-size:150%">En este archivo, se presentará un algoritmo que sirve para simular una cadena de Markov a través de la distribución inicial y la matriz de transición del modelo estocástico. Se tomará como ejemplo el ejercicio visualizado al inicio del semestre, el cual cosnsiste en un tablero dividido en cuatro partes iguales, cada una de las partes tiene un color: rojo, verde, azul y naranja, respectivamente.</p>
<p style="font-family: arial; text-align:justify; font-size:150%">A continuación, se mostrará el código que servirá como ejemplo de la implementación de este algoritmo, consta de una función que nos pide como parámetros tres cosas: la distribución inicial, la matriz de transición y el número de pasos que queremos generar.</p>

```python
import random as rd
def sim(alf, p, m):
#"alf" representa el estado inicial (un vector)
#"p" represeenta la matriz de transicion
#"m" representa el numero de pasos que queremos generar
    alf_acu = []
    acum_alf = 0
    #crea un vector con las sumas acumuladas del estado inicial, para poder determinar cuál será el primer paso
    for prob in alf:
        acum_alf += prob
        alf_acu.append(acum_alf)
    p_acu = []
    #crea un vector de vectores, donde se incluyen las sumas acumuladas por fila de la matriz de transición
    for fila in p:
        fila_acumulada = []
        acum_p = 0
        for prob in fila:
            acum_p += prob
            fila_acumulada.append(acum_p)
        p_acu.append(fila_acumulada)
    
    pasos = []
    num_estados = len(alf)
    r_inicial = rd.random() #número aleatorio (inicial) entre 0 y 1
    #Determina el estado inicial
    for i in range(num_estados):
        if r_inicial <= alf_acu[i]:
            estado_actual = i
            pasos.append(estado_actual)
            break
    #Determina los estados restantes
    for p in range(m - 1):
        r = rd.random()
        fila_actual = p_acu[estado_actual]
        for j in range(num_estados):
            if r <= fila_actual[j]:
                estado_actual = j
                pasos.append(estado_actual)
                break
    return pasos

p = [
    [1/4, 0, 1/4, 2/4], 
    [2/4, 1/4, 0, 1/4], 
    [1/4, 2/4, 1/4, 0], 
    [0, 1/4, 2/4, 1/4]
]
alf = [1/4, 1/4, 1/4, 1/4]
m=5000
resultado = sim(alf, p, m)
print(f"Lista de pasos ({len(resultado)} elementos):")
print(resultado)
na=(sum(1 for n in resultado if n==0))/m
az=(sum(1 for n in resultado if n==1))/m
ve=(sum(1 for n in resultado if n==2))/m
ne=(sum(1 for n in resultado if n==3))/m
print(f"\n Naranja = {na}, Azul = az {az}, Verde = {ve}, Negro = {ne}")
```

<p style="font-family: arial; text-align:justify; font-size:150%">Podemos notar que, con 5000 pasos, es preciso decir que los cuatro colores tienen (aproximadamente) la misma probabilidad de salir (25%), con esto, comprobamos lo que se dijo en clase, al final, los cuatro colores tienen la misma probablidad de aparecer, es decir, se llega a un estado estacionario despues de realizar una cantidad muy grande de pasos.</p>
