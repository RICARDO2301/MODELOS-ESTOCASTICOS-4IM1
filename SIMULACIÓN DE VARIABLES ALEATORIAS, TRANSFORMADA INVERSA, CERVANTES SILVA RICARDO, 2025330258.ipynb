<p style="font-family: arial; text-align:justify; font-size:150%">El método de la transformada inversa nos dice que: Sea \(U\) una variable aleatoria uniforme en \((0,1)\). Para cualquier función de distribución continua \(F\), la variable aleatoria \(X\) definida como \(X = F^{-1}(U)\), donde \(F^{-1}\) es valor de \(x\) tal que \(F(x) = U\), tiene distribución \(F\).</p>

<ul style="font-family: arial; text-align:justify; font-size:150%">
    <li>\(We(\alpha,1)\)<br>
        Recordemos que esta nomenclatura corrresponde a la distribución Weibull, en la cual, la función de distribución es: 
        \[f(x) = \alpha x^{\alpha-1} e^{-x^{-\alpha}}, x \geq 0 \]
    Esto implica: \[F(x) = 1 - e^{-x^{\alpha}}, x \geq 0 \] Aplicando el método de la función inversa:
        \[U = F(x) \]
        \[U = 1 - e^{-x^{\alpha}}\]
        Despenajdo \(x\):
        \[x = (-ln(1-U))^{\frac{1}{a}}\]
        El código que implementa este procedimiento es:
    </li></ul>
  

```python
#Importamos librerías
import numpy as np
import matplotlib.pyplot as plt
import random as rd

#Definición de la fución para generar la muestra
def weibull(a, n):
    muestra=[]
    for i in range(n):
        r=rd.random() #Se genera un número alreatorio uniforme entre (0,1)
        x=(-np.log(1-r))**(1/a) #Se hace la "conversión" pasa que siga la distribución"
        muestra.append(x)
    return muestra

#Se obtiene una muestra de 20000 elementos con alpha=2
mu=weibull(2,20000)
mu_np = np.array(mu)
#Normalización
mu_nor = (mu_np - mu_np.min()) / (mu_np.max() - mu_np.min())

#Se calculan y se muestran la media, varianza y desviación estándar
print("La media es: ", np.mean(mu))
print("La varianza es: ",np.var(mu))
print("La desviación estándar es: ", np.std(mu))

plt.figure()
#Se grafica el historiama respectivo
plt.hist(mu_nor, bins=25, density=True, color="#767F9E")
plt.title("Histograma de la Muestra")

plt.figure()
#Se grafican los datos ordenados para comprobar la función de distribución
plt.plot(np.sort(mu_nor), color='red')
plt.title("Datos Ordenados")

plt.show()
```

<ul style="font-family: arial; text-align:justify; font-size:150%">
    <li>\(Gu(0,1)\)<br>
        Sabemos que esta nomenclatura corrresponde a la distribución de Gumbel, en la cual, la función de distribución es: 
        \[f(x) = e^{-(x+e^{-x})}, -\infty < x < \infty \]
    Esto implica: \[F(x) = e^{-e^{-x}} \] Aplicando el método de la función inversa:
        \[U = F(x) \]
        \[U = e^{-e^{-x}}\]
        Despenajdo \(x\):
        \[x = -ln(-ln(U))\]
        El código que implementa este procedimiento es:
    </li></ul>

```python
#Importamos librerías
import numpy as np
import random as rd
import matplotlib.pyplot as plt

#Definición de la fución para generar la muestra
def gumbel(n):
    muestra=[]
    for i in range(n):
        r=rd.random() #Se genera un número alreatorio uniforme entre (0,1)
        x=-np.log(-np.log(r)) #Se hace la "conversión" pasa que siga la distribución"
        muestra.append(x)
    return muestra

#Se obtiene una muestra de 20000 elementos
mu=gumbel(20000)
#Normalización
mu_norm = (mu - np.min(mu)) / (np.max(mu) - np.min(mu))

#Se calculan y se muestran la media, varianza y desviación estándar
print("La media es: ", np.mean(mu))
print("La varianza es: ",np.var(mu))
print("La desviación estándar es: ", np.std(mu))

plt.figure()
#Se grafica el historiama respectivo
plt.hist(mu_norm, bins=30, density=True, color="#FF4081")
plt.title("Histograma de la Muestra")

plt.figure()
#Se grafican los datos ordenados para comprobar la función de distribución
plt.plot(np.sort(mu_norm), color='purple')
plt.title("Datos ordenados")

plt.show()
```

<ul style="font-family: arial; text-align:justify; font-size:150%">
    <li>\(C(0,1)\)<br>
        Sabemos que esta nomenclatura corresponde a la distribución de Cauchy, en la cual, la función de distribución es: 
        \[f(x) = \frac{1}{\pi(1+x^{2})}, -\infty < x < \infty \]
    Esto implica: \[F(x) = \frac{1}{\pi}arctan(x) + \frac{1}{2} \] Aplicando el método de la función inversa:
        \[U = F(x) \]
        \[U = \frac{1}{\pi}arctan(x) + \frac{1}{2}\]
        Despenajdo \(x\):
        \[x = tan[\pi(U-\frac{1}{2})]\]
        El código que implementa este procedimiento es:
    </li></ul>

```python
#Importamos librerías
import numpy as np
import random as rd
import matplotlib.pyplot as plt

#Definición de la fución para generar la muestra
def cauchy(n):
    muestra=[]
    for i in range(n):
        r=rd.random() #Se genera un número alreatorio uniforme entre (0,1)
        x=np.tan(np.pi*(r-(1/2))) #Se hace la "conversión" pasa que siga la distribución
        muestra.append(x)
    return muestra
#Se obtiene una muestra de 20000 elementos
mu=cauchy(20000)
#Normalización
mu_norm = (mu - np.min(mu)) / (np.max(mu) - np.min(mu))

#Se calculan y se muestran la media, varianza y desviación estándar
print("La media es: ", np.mean(mu))
print("La varianza es: ",np.var(mu))
print("La desviación estándar es: ", np.std(mu))

plt.figure()
#Se grafica el historiama respectivo
plt.hist(mu_norm, bins=30, density=True, color="#FF653F")
plt.title("Histograma de la Muestra")

plt.figure()
#Se grafican los datos ordenados para comprobar la función de distribución
plt.plot(np.sort(mu_norm), color="#831C91")
plt.title("Datos Ordenados")

plt.show()
```

<ul style="font-family: arial; text-align:justify; font-size:150%">
    <li>\(L(0,1)\)<br>
        Sabemos que esta nomenclatura corresponde a la distribución de Laplace, en la cual, la función de distribución es: 
        \[f(x) = \frac{1}{2}e^{-|x|}, -\infty < x < \infty \]
    Esto implica: \[F(x)= \left\{ \begin{array}{lcc}
              \frac{1}{2}e^{x} &   si  & x < 0 \\
              \\ 1-\frac{1}{2}e^{-x} &  si & x \geq 0 \\
              \end{array}
    \right. \] Aplicando el método de la función inversa:
        \[U = F(x) \]
        \[U = \left\{ \begin{array}{lcc}
              \frac{1}{2}e^{x} &   si  & x < 0 \\
              \\ 1-\frac{1}{2}e^{-x} &  si & x \geq 0 \\
              \end{array}
    \right.\]
        Despenajdo \(x\):
        \[x = \left\{ \begin{array}{lcc}
              \ln(2U) &   si  & 0 < U < \frac{1}{2} \\
              \\ -ln(2(1-U)) &  si & \frac{1}{2} \leq U < 1 \\
              \end{array}
    \right.\]
        El código que implementa este procedimiento es:
    </li></ul>

```python
#Importamos librerías
import numpy as np
import random as rd
import matplotlib.pyplot as plt

#Definición de la fución para generar la muestra
def laplace(n):
    muestra=[]
    for i in range(n):
        r=rd.random() #Se genera un número alreatorio uniforme entre (0,1)
        if 0<r<1/2: #Tomando en cuenta las condiciones del despeje, determinamos x
            x=np.log(2*r)
        else:
            x=-np.log(2*(1-r))
        muestra.append(x)
    return muestra

#Se obtiene una muestra de 20000 elementos
mu=laplace(20000)
#Normalización
mu_norm = (mu - np.min(mu)) / (np.max(mu) - np.min(mu))

#Se calculan y se muestran la media, varianza y desviación estándar
print("La media es: ", np.mean(mu))
print("La varianza es: ",np.var(mu))
print("La desviación estándar es: ", np.std(mu))

plt.figure()
#Se grafica el historiama respectivo
plt.hist(mu_norm, bins=30, density=True, color="#F3BE7A")
plt.title("Histograma de la Muestra")

plt.figure()
#Se grafican los datos ordenados para comprobar la función de distribución
plt.plot(np.sort(mu_norm), color="#346739")
plt.title("Datos Ordenados")

plt.show()
```

<ul style="font-family: arial; text-align:justify; font-size:150%">
    <li>\(Par(\alpha)\)<br>
        Sabemos que esta nomenclatura corresponde a la distribución de Pareto, en la cual, la función de distribución es: 
        \[f(x) = \frac{\alpha}{x^{\alpha+1}},  x \geq 1 \]
    Esto implica: \[F(x) = 1 - \frac{1}{x^{\alpha}}, x \geq 1  \] Aplicando el método de la función inversa:
        \[U = F(x) \]
        \[U = 1 - \frac{1}{x^{\alpha}}\]
        Despenajdo \(x\):
        \[x = (1-U)^{-\frac{1}{\alpha}}\]
        El código que implementa este procedimiento es:
    </li></ul>

```python
#Importamos librerías
import numpy as np
import random as rd
import matplotlib.pyplot as plt

#Definición de la fución para generar la muestra
def pareto(a,n):
    muestra=[]
    for i in range(n):
        r=rd.random() #Se genera un número alreatorio uniforme entre (0,1)
        x=(1-r)**(-1/a) #Se hace la "conversión" pasa que siga la distribución
        muestra.append(x)
    return muestra
#Se obtiene una muestra de 20000 elementos con alpha = 3.5
mu=pareto(3.5,20000)
#Normalización
mu_norm = (mu - np.min(mu)) / (np.max(mu) - np.min(mu))

#Se calculan y se muestran la media, varianza y desviación estándar
print("La media es: ", np.mean(mu))
print("La varianza es: ",np.var(mu))
print("La desviación estándar es: ", np.std(mu))

plt.figure()
#Se grafica el histograma respectivo
plt.hist(mu_norm, bins=30, density=True, color="#5E0006")
plt.title("Histograma de la Muestra")

plt.figure()
#Se grafican los datos ordenados para comprobar la función de distribución
plt.plot(np.sort(mu_norm), color="#F77F00")
plt.title("Datos Ordenados")

plt.show()
```
